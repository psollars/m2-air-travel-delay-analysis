```python
import pandas as pd
```

```python
bts_faa_df = pd.read_parquet("./../data/bts_faa_sample_1_42.parquet")

bts_faa_df
```

```python
airport_df = pd.read_csv("./../data/T_MASTER_CORD.csv")

airport_df = airport_df[airport_df["AIRPORT_IS_LATEST"].eq(1)]

airport_df
```

```python
coords = airport_df[
    [
        "AIRPORT",
        "DISPLAY_AIRPORT_NAME",
        "DISPLAY_AIRPORT_CITY_NAME_FULL",
        "LATITUDE",
        "LONGITUDE",
    ]
]

coords
```

```python
origin_coords = coords.add_prefix("Origin_")
bts_faa_coords_df = pd.merge(
    bts_faa_df, origin_coords, left_on="Origin", right_on="Origin_AIRPORT", how="left"
)

dest_coords = coords.add_prefix("Dest_")
bts_faa_coords_df = pd.merge(
    bts_faa_coords_df, dest_coords, left_on="Dest", right_on="Dest_AIRPORT", how="left"
)

bts_faa_coords_df = bts_faa_coords_df.drop(
    ["Origin_AIRPORT", "Dest_AIRPORT"], axis="columns"
)
```
