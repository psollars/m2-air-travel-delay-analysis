```python
dereg_2023 = pd.read_csv("./../data/ReleasableAircraft.2023/DEREG_cleaned.txt", delimiter=",")
```

```python
dereg_2023.dropna(axis=1, how="all", inplace=True)

dereg_2023.columns
```

```python
dereg_2023["CANCEL-DATE"] = pd.to_datetime(dereg_2023["CANCEL-DATE"], format="%Y%m%d", errors="coerce")

dereg_2023_filtered = dereg_2023[dereg_2023['CANCEL-DATE'].dt.year.ge(2015) | dereg_2023['CANCEL-DATE'].isnull()]

dereg_2023_filtered
```

```python
with sqlite3.connect("dereg_2023_filtered.db") as conn:
    dereg_2023_filtered.to_sql(name="dereg_2023_filtered", con=conn, if_exists="replace", index=False)
```
